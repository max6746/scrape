```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://lightning.nsstc.nasa.gov/isslisib/tmp/lisflashes-61.3.70.109.txt'
# Create object page
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
lines = soup.text.splitlines()
# lines
```

```python
data = []
for line in lines[1:]:
    # cleaning data
    element = []
    line = line.replace('(', '')
    line = line.replace(',', '')
    line = line.replace(')', '')
    firstelm = line.split(']')
    element.append(firstelm[0] + ']')
    element = element + firstelm[1].split()
    data.append(element)

data
```
